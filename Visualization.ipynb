```python
# import basic libraries
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

```python
# seaborn has a inbuilt dataset
sns.get_dataset_names()
```

```python
# load iris data
data=sns.load_dataset("iris")
```

```python
data
```

# Relational Plots

Relational plots are used to understand relationship between two variables.

* Line plot
* Scatter plot
* Replot

```python
# line plot
sns.lineplot(data=data,x="sepal_length",y="sepal_width")
```

```python
# line plot
sns.lineplot(data=data,x="sepal_length",y="sepal_width",color="green")
```

```python
sns.lineplot(data=data,x='sepal_length',y='sepal_width',hue='species')
```

```python
sns.lineplot(data=data,x='sepal_length',y='sepal_width',hue='species',ci=None)
```

# Scatter

```python
sns.scatterplot(data=data,x='petal_width',y='petal_length')
plt.show
```

```python
sns.scatterplot(data=data,x='petal_width',y='petal_length',hue='species')
plt.show
```

```python
sns.scatterplot(data=data,x='petal_width',y='petal_length',hue='species',palette='rainbow')
plt.show
```

# Rel Plot

relplot by default it is scatter plot

relplot can be converted into line plot by giving kind=line

```python

sns.relplot(data=data,x='petal_length',y='sepal_width')
```

```python
sns.relplot(data=data,x='petal_length',y='sepal_width',kind='line')
```

```python
sns.relplot(data=data,x='petal_length',y='sepal_width',kind='line',ci=None)
```
